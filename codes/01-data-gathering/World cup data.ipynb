```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../../PROject DATA/fifa_ranking.csv')
df= df .loc[:,['rank', 'country_full', 'country_abrv', 'cur_year_avg_weighted', 'rank_date','two_year_ago_weighted', 'three_year_ago_weighted']]
df=df.replace({"IR Iran": "Iran"})
df
```

```python
df['weighted_points'] = df['cur_year_avg_weighted'] + df['two_year_ago_weighted'] + df['three_year_ago_weighted']
df['rank_date'] = pd.to_datetime(df['rank_date'])

matches = pd.read_csv('../../PROject DATA/results.csv')
matches =  matches.replace({'Germany DR': 'Germany', 'China': 'China PR'})
matches['date'] = pd.to_datetime(matches['date'])
```

```python
world_cup = pd.read_csv('../../PROject DATA/WC 2022 .csv')
world_cup = world_cup.loc[:, ['Team', 'Group', 'First match \nagainst', 'Second match\n against', 'Third match\n against']]
world_cup = world_cup.dropna(how='all')
world_cup = world_cup.replace({"IRAN": "Iran", "Costarica": "Costa Rica", "Porugal": "Portugal", "Korea" : "Korea Republic"})
```

```python
matches=matches[matches['tournament']!='Friendly']
matches.shape
```

```python
import datetime

df['year']=df['rank_date'].dt.year
matches['year']=matches['date'].dt.year

df_sub=df[df['rank_date']>'2011-08-24']
matches_sub = matches[matches['date']>'2011-08-24']
```

```python
df_sub=df_sub[['rank','country_full','weighted_points','year']]
df_sub=df_sub.groupby(['year','country_full'],as_index=False).agg({'rank':'mean','weighted_points':'mean',})

matches_sub
```

```python
matches_all= matches_sub.merge(df_sub,left_on=['home_team','year'], right_on=['country_full', 'year'], how='inner')
matches_all = matches_all.merge(df_sub, left_on=['away_team', 'year'], right_on=['country_full', 'year'], how='inner')
matches_all.drop(['date', 'tournament', 'city', 'country', 'year', 'country_full_x', 'country_full_y'], axis=1, inplace=True)
matches_all
	
```

```python
matches_all['weighted_diff'] = matches_all['weighted_points_x'] - matches_all['weighted_points_y']
matches_all['rank_diff'] = matches_all['rank_x'] - matches_all['rank_y']
matches_all['neutral'] = matches_all['neutral'].astype(int)
matches_all['is_win'] = (matches_all['home_score'] - matches_all['away_score']).apply(lambda x: 1 if x>0 else 0)
train = matches_all.drop(['home_team', 'away_team', 'home_score', 'away_score'], axis=1)
train=train.rename(columns={'is_win':'label'})
train.to_csv('WC cleaned data.csv')
train
```
